```python
# --- Strip column names to remove extra spaces ---
for df in [df_customers, df_customer_reviews, df_customer_journey, df_engagement, df_products, df_geography]:
    df.columns = df.columns.str.strip()

# --- Merge DataFrames ---
merged_df = (df_customers
             .merge(df_customer_reviews, on="CustomerID", how="left")
             .merge(df_customer_journey, on=["CustomerID", "ProductID"], how="left")
             .merge(df_engagement, on="ProductID", how="left")
             .merge(df_products, on="ProductID", how="left")
             .merge(df_geography, on="GeographyID", how="left"))

# ✅ Replace NaN, NaT, and Empty Strings with None (MySQL NULL compatibility)
merged_df = merged_df.replace({np.nan: None, pd.NA: None, "nan": None, "": None}).where(pd.notnull(merged_df), None)

# ✅ Convert Date Columns to Proper Format
date_columns = ["ReviewDate", "VisitDate", "EngagementDate"]
for col in date_columns:
    if col in merged_df.columns:
        merged_df[col] = pd.to_datetime(merged_df[col], errors="coerce").dt.strftime('%Y-%m-%d')

# --- Drop Table if it Exists ---
cursor.execute("DROP TABLE IF EXISTS merged_data")

# --- Create Table with AUTO_INCREMENT Primary Key ---
create_table_query = """
CREATE TABLE merged_data (
    id INT AUTO_INCREMENT PRIMARY KEY,
    CustomerID INT,
    CustomerName VARCHAR(255),
    Email VARCHAR(255),
    Gender ENUM('Male', 'Female', 'Other'),
    Age INT,
    GeographyID INT,
    Country VARCHAR(100),
    City VARCHAR(100),
    ReviewID INT,
    ReviewDate DATE,
    Rating INT CHECK (Rating BETWEEN 1 AND 5),
    ReviewText TEXT,
    JourneyID INT,
    VisitDate DATE,
    Stage VARCHAR(100),
    Action VARCHAR(255),
    Duration INT,
    ProductID INT,
    ProductName VARCHAR(255),
    Category VARCHAR(100),
    Price DECIMAL(10,2),
    ContentID INT,
    ContentType VARCHAR(100),
    Likes INT,
    EngagementDate DATE,
    CampaignID INT,
    Views INT,
    Clicks INT
);
"""
cursor.execute(create_table_query)
print("✅ Table 'merged_data' created successfully!")

# --- Prepare Insert Statement ---
insert_query = """
    INSERT INTO merged_data (CustomerID, CustomerName, Email, Gender, Age, GeographyID, Country, City,
                             ReviewID, ReviewDate, Rating, ReviewText, 
                             JourneyID, VisitDate, Stage, Action, Duration,
                             ProductID, ProductName, Category, Price,
                             ContentID, ContentType, Likes, EngagementDate, CampaignID, Views, Clicks) 
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, 
            %s, %s, %s, %s, 
            %s, %s, %s, %s, %s,
            %s, %s, %s, %s, 
            %s, %s, %s, %s, %s, %s, %s)
"""

# --- Select Only the Required Columns for Insertion ---
cols_to_insert = [
    "CustomerID", "CustomerName", "Email", "Gender", "Age", "GeographyID", "Country", "City",
    "ReviewID", "ReviewDate", "Rating", "ReviewText", 
    "JourneyID", "VisitDate", "Stage", "Action", "Duration",
    "ProductID", "ProductName", "Category", "Price",
    "ContentID", "ContentType", "Likes", "EngagementDate", "CampaignID", "Views", "Clicks"
]
merged_df_subset = merged_df[cols_to_insert]

# ✅ Debugging: Print Sample Data Before Insertion
print("Sample row for insertion:", merged_df_subset.iloc[0].to_dict())

# ✅ Convert DataFrame rows to a list of tuples for insertion
data_tuples = [
    tuple(None if (isinstance(value, float) and np.isnan(value)) else value for value in row)
    for row in merged_df_subset.itertuples(index=False, name=None)
]

# --- Insert Merged Data Using executemany ---
try:
    cursor.executemany(insert_query, data_tuples)
    conn.commit()
    print("✅ Merged CSV data inserted successfully!") 
except db.Error as err:
    print(f"❌ Error inserting data: {err}")
```

```python
# customers who brought multiple products 

multiple_products_query = """
SELECT 
    m.CustomerID,
    m.CustomerName,
    GROUP_CONCAT(DISTINCT p.ProductName ORDER BY p.ProductName SEPARATOR ', ') AS ProductsBought,
    COUNT(DISTINCT m.ProductID) AS TotalProductsBought
FROM merged_data m
JOIN products p ON m.ProductID = p.ProductID
WHERE m.ProductID IS NOT NULL
GROUP BY m.CustomerID, m.CustomerName
HAVING COUNT(DISTINCT m.ProductID) > 1
ORDER BY TotalProductsBought DESC;
"""

# ✅ Execute Query
cursor.execute(multiple_products_query)
multiple_products = cursor.fetchall()

# ✅ Convert to DataFrame
multiple_products_data = pd.DataFrame(multiple_products, columns=cursor.column_names)
multiple_products_data
```

```python
#Customers who viewed the product but not purchased  

viewed_not_purchased_query = """
SELECT 
    cj.CustomerID,
    c.CustomerName,
    c.Email,
    c.Gender,
    c.Age,
    c.GeographyID,
    cj.ProductID,
    p.ProductName,
    cj.Action
FROM customer_journey cj
JOIN customers c ON cj.CustomerID = c.CustomerID
JOIN products p ON cj.ProductID = p.ProductID
WHERE cj.Action = 'View'
AND NOT EXISTS (
    SELECT 1 
    FROM customer_journey cj2
    WHERE cj.CustomerID = cj2.CustomerID
    AND cj.ProductID = cj2.ProductID
    AND cj2.Action = 'Purchase'
)
ORDER BY cj.CustomerID, cj.ProductID;
"""

# ✅ Execute Query
cursor.execute(viewed_not_purchased_query)
viewed_not_purchased = cursor.fetchall()

# ✅ Convert to DataFrame
viewed_not_purchased_data = pd.DataFrame(viewed_not_purchased, columns=cursor.column_names)
viewed_not_purchased_data
```

```python
#why customers dropped off ?

drop_off_query = """
SELECT 
    Stage, 
    COUNT(*) AS DropOffCount 
FROM customer_journey 
WHERE Action = 'Drop-off'
GROUP BY Stage
ORDER BY DropOffCount DESC;
"""

cursor.execute(drop_off_query)
dropoff_data = cursor.fetchall()
dropoffrate_data = pd.DataFrame(dropoff_data, columns=cursor.column_names)
dropoffrate_data
```

```python
#Finding the drop off rate 

drop_offrate="""
WITH stage_counts AS (
    SELECT 
        cj.Stage,
        COUNT(DISTINCT cj.CustomerID) AS total_customers,
        COUNT(DISTINCT CASE WHEN cj.Action = 'Drop-off' THEN cj.CustomerID END) AS dropoff_customers
    FROM customer_journey cj
    JOIN products p ON cj.ProductID = p.ProductID
    WHERE p.Category = 'Sports'
    GROUP BY cj.Stage
)
SELECT 
    Stage, 
    total_customers,
    dropoff_customers,
    (dropoff_customers * 100.0 / total_customers) AS dropoff_rate
FROM stage_counts
ORDER BY dropoff_rate DESC;
"""
cursor.execute(drop_offrate)
dropoffrate = cursor.fetchall()
dropoffrate_data = pd.DataFrame(dropoffrate,columns=cursor.column_names )
dropoffrate_data
```

```python
# Action leadind to succes conversions 

purchase_view_query = """
SELECT 
    Stage,
    Action,
    COUNT(*) AS Occurrences,
    ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(), 2) AS Percentage
FROM customer_journey
WHERE Action IN ('View', 'Purchase','click', 'Drop off')
AND Stage IN ('Homepage', 'ProductPage', 'Checkout')
GROUP BY Stage, Action
ORDER BY Percentage DESC;
"""

# ✅ Execute Query
cursor.execute(purchase_view_query)
purchase_view = cursor.fetchall()

# ✅ Convert to DataFrame
purchase_view_data = pd.DataFrame(purchase_view, columns=cursor.column_names)
purchase_view_data
```

```python
#Average duration per stage

avg_duration_query = """
SELECT 
    Stage,
    ROUND(AVG(Duration), 2) AS AvgDuration
FROM customer_journey
WHERE Stage IN ('Checkout', 'ProductPage', 'Homepage') 
      AND Duration IS NOT NULL
GROUP BY Stage
ORDER BY AvgDuration DESC;
"""

# ✅ Execute Query
cursor.execute(avg_duration_query)
avg_duration = cursor.fetchall()

# ✅ Convert to DataFrame
avg_duration_data = pd.DataFrame(avg_duration, columns=cursor.column_names)
avg_duration_data
```

```python
#Repeat vs First Time buyers 

sql = """
WITH PurchaseCounts AS (
    SELECT 
        c.CustomerID,
        c.CustomerName,
        COUNT(DISTINCT cr.ReviewID) AS purchase_count
    FROM customer_reviews cr
    JOIN customers c ON cr.CustomerID = c.CustomerID
    GROUP BY c.CustomerID, c.CustomerName
)
SELECT 
    SUM(CASE WHEN purchase_count = 1 THEN 1 ELSE 0 END) AS first_time_buyers,
    SUM(CASE WHEN purchase_count > 1 THEN 1 ELSE 0 END) AS repeat_buyers,
    COUNT(CustomerID) AS total_buyers
FROM PurchaseCounts;
"""

# ✅ Execute Query
cursor.execute(sql)
repeat_vs_first_time = cursor.fetchall()

# ✅ Convert to DataFrame
repeat_vs_first_time_data = pd.DataFrame(repeat_vs_first_time, columns=cursor.column_names)

# ✅ Display Data
repeat_vs_first_time_data
```

```python
#Repeat vs First Time buyers 

sql = """
WITH PurchaseCounts AS (
    SELECT 
        c.CustomerID,
        c.CustomerName,
        COUNT(DISTINCT cr.ReviewID) AS purchase_count,
        COUNT(DISTINCT cr.ProductID) AS total_products_bought
    FROM customer_reviews cr
    JOIN customers c ON cr.CustomerID = c.CustomerID
    GROUP BY c.CustomerID, c.CustomerName
),
ProductDetails AS (
    SELECT 
        cr.CustomerID,
        GROUP_CONCAT(DISTINCT p.ProductName ORDER BY p.ProductName ASC SEPARATOR ', ') AS ProductsBought,
        GROUP_CONCAT(DISTINCT p.Category ORDER BY p.Category ASC SEPARATOR ', ') AS CategoriesBought
    FROM customer_reviews cr
    JOIN products p ON cr.ProductID = p.ProductID
    GROUP BY cr.CustomerID
)
SELECT 
    pc.CustomerID,
    pc.CustomerName,
    pc.total_products_bought,
    pd.ProductsBought,
    pd.CategoriesBought,
    CASE 
        WHEN pc.purchase_count = 1 THEN 'First-Time Buyer'
        ELSE 'Repeat Buyer'
    END AS BuyerType
FROM PurchaseCounts pc
LEFT JOIN ProductDetails pd ON pc.CustomerID = pd.CustomerID
ORDER BY BuyerType DESC, pc.total_products_bought DESC;
"""

# ✅ Execute Query
cursor.execute(sql)
repeat_vs_first_time = cursor.fetchall()

# ✅ Convert to DataFrame
repeat_vs_first_time_data = pd.DataFrame(repeat_vs_first_time, columns=cursor.column_names)

# ✅ Display Data
repeat_vs_first_time_data
```

```python
retention_rate_sql = """
WITH PurchaseCounts AS (
    SELECT 
        c.CustomerID,
        COUNT(DISTINCT cr.ReviewID) AS purchase_count
    FROM customer_reviews cr
    JOIN customers c ON cr.CustomerID = c.CustomerID
    GROUP BY c.CustomerID
)
SELECT 
    COUNT(CASE WHEN purchase_count > 1 THEN 1 END) AS repeat_buyers,
    COUNT(CustomerID) AS total_buyers,
    ROUND((COUNT(CASE WHEN purchase_count > 1 THEN 1 END) / COUNT(CustomerID)) * 100, 2) AS retention_rate
FROM PurchaseCounts;
"""

# ✅ Execute Query
cursor.execute(retention_rate_sql)
retention_result = cursor.fetchall()

# ✅ Convert to DataFrame
retention_data = pd.DataFrame(retention_result, columns=cursor.column_names)

# ✅ Display Data
retention_data
```

```python
#Best performing products per region according to rating and reviews 

best_products_query = """
WITH ProductPerformance AS (
    SELECT 
        GeographyID,
        Country,
        City,
        ProductID,
        ProductName,
        COUNT(ReviewID) AS TotalReviews,  -- Count of reviews per product
        ROUND(AVG(Rating), 2) AS AvgRating,  -- Average rating per product
        RANK() OVER(PARTITION BY GeographyID ORDER BY COUNT(ReviewID) DESC, AVG(Rating) DESC) AS ProductRank
    FROM merged_data
    WHERE ProductID IS NOT NULL AND Rating IS NOT NULL
    GROUP BY GeographyID, Country, City, ProductID, ProductName
)
SELECT 
    GeographyID,
    Country,
    City,
    ProductID,
    ProductName,
    TotalReviews,
    AvgRating
FROM ProductPerformance
WHERE ProductRank = 1;  -- Select the top-performing product per region
"""

# ✅ Execute Query
cursor.execute(best_products_query)
best_products = cursor.fetchall()

# ✅ Convert to DataFrame
best_products_data = pd.DataFrame(best_products, columns=cursor.column_names)

# ✅ Display Data
best_products_data

```
